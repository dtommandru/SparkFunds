# Data Cleaning

### We can see that permalink of Companies data frame and Companies_permalink of rounds dataframe are different although they are same
### Let's convert them to lower case

```python
companies["permalink"]=companies["permalink"].str.lower()
rounds["company_permalink"]=rounds["company_permalink"].str.lower()
```

```python
len(companies["permalink"].unique())
```

```python
len(rounds["company_permalink"].unique())
```

Appears like 2 extra companies are there in rounds file which received investment while not listed in companies file let's check them out 

```python
rounds.loc[~rounds["company_permalink"].isin(companies["permalink"])]
```

Seems like problemiswith encoding.By converting into english

```python
rounds['company_permalink'] = rounds.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
rounds.loc[~rounds['company_permalink'].isin(companies['permalink']), :]
```

```python
companies['permalink'] = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
```

```python
df.groupby("funding_round_type")["raised_amount_usd"].median()/1000000
```

## Country Wise Analysis

```python
df = df[df.funding_round_type=="venture"]

# group by country codes and compare the total funding amounts
country_wise = df.groupby('country_code')['raised_amount_usd'].sum().sort_values(ascending=False)
print(country_wise)
```

```python
top9 = country_wise[0:10]
top9
```

#### Top3 English Speaking Countries : "USA","GBR","IND"

```python
top3_eng = top9.loc[["USA","GBR","IND"]]
top3_eng.sort_values(ascending=False)
```
